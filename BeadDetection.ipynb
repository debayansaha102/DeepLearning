```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile
from scipy.ndimage.filters import gaussian_filter
from skimage.measure import *
from skimage.morphology import binary_erosion
import glob
%gui qt5
```

```python
folder = "D://Data//2018-08-24-13-17-42-13-RandomZernike//stacks"
Zrange = 80
XYrange = 64
```

```python
files = glob.glob(folder+"//Images/*.tif")
zr = Zrange//2
xy = XYrange//2
```

```python
timepoint = [f[f.rfind('\\')+1:f.find('.tif')] for f in files]
```

```python
for t,f in enumerate(files):
    print(t)
    img = tifffile.imread(f)
    img1 = gaussian_filter(img,4)
    img1 = gaussian_filter(img,4)
    mask = img1> np.percentile(img1,99.9)
    img2 = binary_erosion(mask)
    img3 = label(img2,8)
    prop = regionprops(img3,img1) 
    pos = []
    for i in prop:
        (z,r,c) = i.centroid
        if (z-zr)<=0 or (z+zr)>=img.shape[0]:
            continue
        if (r-xy)<=0:
            continue
        if (r+xy)>=img.shape[1]:
            continue;
        if (c-xy)<=0:
            continue
        if (c+xy)>=img.shape[2]:
            continue;
        pos.append((z,r,c))
    for i,p in enumerate(pos):
        tifffile.imsave(folder+"\\Beads\\%s_%d"%(timepoint[t],i),img[int(p[0]-zr):int(p[0]+zr),int(p[1]-xy):int(p[1]+xy),int(p[2]-xy):int(p[2]+xy)])
    #break
```
